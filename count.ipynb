```python
# Iterate all tables

import glob
import os

csv_files = glob.glob("csv/*.csv")

RIVALS = ["Keming", "Professor Yuan", "Haoru", "David", "Vic", "Carl", "Ishun", "Hitokiri Joe", "Natsuhisa", "Bo Kang", "Wataru", "Jud", "Alex Sebastian", "Ballin", "Einar", "Jesse"]
MALES = ["Zhanwu", "Fan Jun", "Changping", "Linus", "Will", "Jeff", "Mohei", "Nishida", "Haruji", "Jin Zhang", "Moritoo", "Curly", "Devlin", "Johnny", "Eric", "Lewt"]
LADIES = ["Shuzhen", "Fang Qing", "Suhua", "Sabrina", "Barbara", "Vicki", "Osan", "Shizuko", "Eri", "Ye Suping", "Kesa", "Laurey", "Alicia", "Gilda", "Morgana", "Pearl"]

t1 = 0
t2 = 0
u = 0

for csv_file in csv_files:
	allids = []
	table_name = os.path.splitext(os.path.basename(csv_file))[0]
	print(f"Table: {table_name}")
	lad = 0
	lady = ""
	rival = ""
	male = ""
	riv = 0
	mal = 0
	with open(csv_file, "r") as f:
		l = 0
		t = 0
		n = 0
		for line in f:
			# skip header 2 lines
			if l < 2:
				l += 1
				continue
			#print(line.strip().split(",")[5])
			s = line.strip().split(",")[5]
			s = s[1:-1]
			if s != "N/A":
				# check if id is unique
				if s not in allids:
					allids.append(s)
			if True:
				#print(line.strip())
				t += 1
				t1 += 1
				if s in RIVALS or s in MALES or s in LADIES:
					n += 1
					t2 += 1
					if s in RIVALS:
						rival = s

						riv += 1
					elif s in MALES:
						male = s

						mal += 1
					elif s in LADIES:
						lad += 1

						lady = s
				
		u += len(allids)
		print(f"Total unique: {len(allids)}")
		print(f"Total: {t}")
		print(f"Total with names: {n}")
		print(f"Total rivals: {riv} ({rival})")
		print(f"Total males: {mal} ({male})")
		print(f"Total females: {lad} ({lady})")
	print()
	print()

print(f"Total: {t1}")
print(f"Total unique: {u}")
print(f"Total with names: {t2}")
```
